#TASK-1.Customer Email Classification




```python
def classify_customer_email(email):
    categories = ["Complaint", "Inquiry", "Feedback", "Support Request"]
    prompt = f"Classify the following email into one of {categories}:\n\nEmail: {email}"
    return prompt


print(classify_customer_email("I have not received my order yet."))
```

#TASK.2 – Intent Classification for Chatbot Queries

```python
def classify_intent(query):
    intents = ["Book Flight", "Cancel Ticket", "Track Order", "General Inquiry"]
    prompt = f"""
Classify the user query into one of the following intents:
{intents}

User Query: "{query}"
Intent:
"""
    return prompt


print(classify_intent("I want to cancel my booking."))
```
